```python
import requests
import json

# Define the OpenRouter API endpoint (replace with the correct URL from OpenRouter documentation)
api_url = 'https://openrouter.ai/api/v1/chat/completions'  # Example URL; replace with actual one

# Your OpenRouter API Key (replace with your actual API key)
api_key = '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'

# Headers for the request
headers = {
    'Authorization': f'Bearer {api_key}',
    'Content-Type': 'application/json'
}

# Define a list of different prompts to experiment with
prompts = [
    "What is the capital of France?",
    "Can you write a short story about a dragon and a knight?",
    "Explain quantum physics in simple terms.",
    "What is the meaning of life?",
    "Write a poem about nature."
]

# Function to make the API request and get a response
def get_openrouter_response(prompt):
    payload = {
        'model': 'gpt-3.5-turbo',  # Or whichever model you want to use (e.g., 'gpt-4')
        'prompt': prompt,
        'max_tokens': 100,
        'temperature': 0.7
    }
    
    response = requests.post(api_url, headers=headers, data=json.dumps(payload))
    
    if response.status_code == 200:
        return response.json()['choices'][0]['text'].strip()
    else:
        return f"Error: {response.status_code}, {response.text}"

# Try each prompt and print the response
for prompt in prompts:
    print(f"Prompt: {prompt}")
    response = get_openrouter_response(prompt)
    print(f"Response: {response}\n")
```
