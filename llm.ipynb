```python
import requests
import json

class Chatbot:
    def __init__(self, api_key, model="openai/gpt-3.5-turbo"):
        self.api_key = api_key
        self.model = model

    def send_message(self, prompt):
        url = "https://openrouter.ai/api/v1/chat/completions"
        headers = {"Authorization": f"Bearer {self.api_key}"}
        data = {
            "model": self.model,
            "messages": [{"role": "user", "content": prompt}]
        }
        response = requests.post(url, headers=headers, json=data)
        if response.status_code == 200:
            reply = response.json()["choices"][0]["message"]["content"]
            return reply.strip()
        else:
            raise Exception(f"Error: {response.status_code} - {response.text}")

api_key = "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"
chatbot = Chatbot(api_key)

user_input = "What is sentiment analysis?"
response = chatbot.send_message(user_input)
print(response)
```

```python
class SentimentAnalyzer:
    def __init__(self, api_key, model="openai/gpt-3.5-turbo"):
        self.api_key = api_key
        self.model = model

    def analyze_sentiment(self, text):
        prompt = f"Analyze the sentiment of the following text: \"{text}\". Respond with 'Positive', 'Negative', or 'Neutral'."
        return chatbot.send_message(prompt)

api_key = "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"
sentiment_analyzer = SentimentAnalyzer(api_key)

text = "I love how this tool simplifies my tasks!"
sentiment = sentiment_analyzer.analyze_sentiment(text)
print(sentiment)
```
